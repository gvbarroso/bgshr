```python
import time
import bgshr
import scipy
import numpy as np
import pandas
import matplotlib.pylab as plt, matplotlib


# Define paths to data
lookup_tbl_file = "lookup_tables/lookup_tbl_equilibrium.csv.gz"
rec_map_file = "data/YRI_recombination_map_hapmap_format_hg38_chr_22.txt.gz"
mut_tbl_file = "data/roulette_tbl_chr22.csv.gz"
exons_file = "data/exons_chr22.bed.gz"
```

```python
# Definite parameters
Ne = 27000
# DFE parameters inferred by Kim et al. 2017
dfe_shape = 0.215
dfe_scale = 0.0240
dfe_p_neu = 0.302
```

```python
# Define the length of the chromosome and the grid of focal sites
# We use a coarse 10kb grid here to speed up predictions
start = 0
L = 51000000
spacing = 10000
xs = np.arange(start, L, spacing)
```

```python
# Load and extend the lookup table using classic BGS theory
df, Ne_df = bgshr.Predict.load_extended_lookup_table(lookup_tbl_file, num_s_vals=16)
print(np.unique(df["r"]))

# Splines allow interpolation in `B` between `r` values from the lookup table
u_vals, s_vals, splines = bgshr.Util.generate_cubic_splines(df)
print(u_vals)
print(s_vals)
```

```python
# We scale the parameters `u`, `r` and `s`. This is an approximate way of making predictions 
# for a different Ne than that embodied in the lookup table
scaling = Ne / Ne_df
print(scaling)

# Load recombination map and apply scaling
rec_map = bgshr.Util.load_recombination_map(rec_map_file, L, scaling=scaling)
```

```python
# Load the mutation map 
mut_df = pandas.read_csv(mut_tbl_file)

# Here we use a table to construct an approximate site-resolution map, but in
# practice we load a site-resolution map directly
columns = mut_df.columns
tbl_intervals = np.stack((mut_df[columns[1]], 
    mut_df[columns[2]]), axis=1).astype(np.int64)
values = np.array(mut_df[columns[4]])
mut_map = bgshr.Util.construct_site_map(tbl_intervals, values)

# Load elements
elements, _ = bgshr.Util.read_bedfile(exons_file)


# From this, we compute average `u` in each element
elem_lens, elem_muts = bgshr.Util.compute_windowed_average(elements, mut_map)
print(np.sum(elem_lens))
print(elem_muts)

# Scaling is applied to the mutation rate
elem_muts = scaling * elem_muts
print(elem_muts)
```

```python
# `s` is scaled by increasing the scale parameter of the DFE
scaled_dfe_scale = scaling * dfe_scale
print(scaled_dfe_scale)
dfe_params = {"type": "gamma_neutral", "shape": dfe_shape,
              "scale": scaled_dfe_scale, "p_neu": dfe_p_neu}
weights = bgshr.Util._get_dfe_weights(dfe_params, s_vals)
print(weights)
```

```python
# The zeroth round of B prediction
Bs = bgshr.Predict.Bvals(
    xs,
    s_vals[:-1],
    splines,
    u=elem_muts,
    L=L,
    rmap=rec_map,
    elements=elements,
    max_r=0.5
)
B = bgshr.Util.integrate_with_weights(Bs, weights[:-1])
print(B)
```

```python
# A single iteration of the interference correction, which scales local parameters
# by a prior B map
B_map = scipy.interpolate.CubicSpline(xs, B, bc_type="natural")
rec_map_adjusted = bgshr.Util.adjust_recombination_map(rec_map, B_map)
Bs = bgshr.Predict.Bvals(
    xs,
    s_vals[:-1],
    splines,
    u=elem_muts,
    L=L,
    rmap=rec_map_adjusted,
    elements=elements,
    max_r=0.5,
    bmap=B_map
)
B1 = bgshr.Util.integrate_with_weights(Bs, weights[:-1])
print(B1)
```

```python
fig, ax = plt.subplots(figsize=(11, 4.5), layout="constrained")
ax.plot(xs, B, label="naive prediction")
ax.plot(xs, B1, label="interference correction 1")
ax.legend(framealpha=0)
ax.set_xlim(0, L)
ax.set_title("Diversity reduction on chromosome 22 (exon model)")
ax.set_ylabel("Diversity reduction ($B$)")
ax.set_xlabel("Position (Mb)")
ax.set_xticks([0, 1e7, 2e7, 3e7, 4e7, 5e7], [0, 10, 20, 30, 40, 50])
plt.savefig("figures/chr22_10kb_exonmodel.pdf")
```
